```python
# Twitterのフォロワーを取得するためのスクリプト
# coding:utf-8
import tweepy
import pandas as pd
import time
from tqdm import tqdm
# import os

# Oauth
#CONSUMER_KEY = os.environ['CONSUMER_KEY']
#CONSUMER_SECRET = os.environ['CONSUMER_SECRET']
#ACCESS_TOKEN = os.environ['ACCESS_TOKEN']
#ACCESS_TOKEN_SECRET = os.environ['ACCESS_TOKEN_SECRET']

# ↓Twitterの認証情報を入力する
CONSUMER_KEY = ''
CONSUMER_SECRET = ''
ACCESS_TOKEN = ''
ACCESS_TOKEN_SECRET = ''

# save dataframe
columns = ['screen_name', 'username', 'description', 'follower', 'following']
df = pd.DataFrame(columns=columns)

cursor = -1
while cursor != 0:
   auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
   auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
   api = tweepy.API(auth, wait_on_rate_limit=True)
   # ↓取得したいTwitterユーザーのscreen_nameを入力する
   itr = tweepy.Cursor(api.followers_ids, id='#####', cursor=cursor).pages()
   try:
       for follower_id in itr.next():
           try:
                user = api.get_user(follower_id)
                screen_name = user.screen_name
                username = user.name
                description = user.description
                follower = user.followers_count
                following = user.friends_count
                se = pd.Series([screen_name, username, description, follower, following], columns)
                df = df.append(se, ignore_index=True)
                print(df)
           except tweepy.error.TweepError as e:
               print(e.reason)
   except ConnectionError as e:
       print(e.reason)
   cursor = itr.next_cursor

# df.head()

# ↓CSVデータを出力、ファイル名を入力する（拡張子は.csv）
# export CSV
filename = '#####.csv'
df.to_csv(filename, encoding = 'utf-8-sig')
```
